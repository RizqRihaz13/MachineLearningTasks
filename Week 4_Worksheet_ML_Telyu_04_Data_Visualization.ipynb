Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Tentukan path file di Google Drive
file_path = '/content/drive/MyDrive/StudentsPerformance.csv'

# Baca dataset sebagai DataFrame
df = pd.read_csv(file_path)

# Fungsi untuk membuat pie chart dari kolom kategorikal
def plot_pie_chart(column_name):
    data = df[column_name].value_counts()
    plt.figure(figsize=(8, 6))
    data.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab20', legend=False)
    plt.title(f'Percentage of {column_name}')
    plt.ylabel('')  # Menghilangkan label default di sumbu Y
    plt.axis('equal')  # Membuat pie chart berbentuk lingkaran
    plt.show()

# Contoh penggunaan untuk kolom 'gender'
plot_pie_chart('gender')

# Anda bisa mencoba untuk kolom lain seperti:
# plot_pie_chart('race/ethnicity')
# plot_pie_chart('parental level of education')
# plot_pie_chart('lunch')
# plot_pie_chart('test preparation course')
````

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset
df = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')

# Set tema untuk seaborn agar tampilan lebih estetik
sns.set_theme(style="whitegrid")

# Membuat subplots untuk setiap variabel dan nilai (math, reading, writing)
def plot_score_distribution():
    plt.figure(figsize=(18, 12))

    # 1. Gender vs Math, Reading, Writing Scores
    plt.subplot(3, 2, 1)
    sns.boxplot(x='gender', y='math score', data=df, palette='coolwarm')
    plt.title('Math Score by Gender')

    plt.subplot(3, 2, 2)
    sns.boxplot(x='gender', y='reading score', data=df, palette='coolwarm')
    plt.title('Reading Score by Gender')

    plt.subplot(3, 2, 3)
    sns.boxplot(x='gender', y='writing score', data=df, palette='coolwarm')
    plt.title('Writing Score by Gender')

    # 2. Race/Ethnicity vs Math, Reading, Writing Scores
    plt.subplot(3, 2, 4)
    sns.boxplot(x='race/ethnicity', y='math score', data=df, palette='viridis')
    plt.title('Math Score by Race/Ethnicity')

    plt.subplot(3, 2, 5)
    sns.boxplot(x='race/ethnicity', y='reading score', data=df, palette='viridis')
    plt.title('Reading Score by Race/Ethnicity')

    plt.subplot(3, 2, 6)
    sns.boxplot(x='race/ethnicity', y='writing score', data=df, palette='viridis')
    plt.title('Writing Score by Race/Ethnicity')

    plt.tight_layout()
    plt.show()

    # 3. Parental Level of Education vs Average Score
    plt.figure(figsize=(18, 8))
    sns.barplot(x='parental level of education', y=df[['math score', 'reading score', 'writing score']].mean(axis=1),
                data=df, palette='pastel')
    plt.xticks(rotation=45)
    plt.title('Average Scores by Parental Level of Education')
    plt.xlabel('Parental Level of Education')
    plt.ylabel('Average Score')
    plt.tight_layout()
    plt.show()

    # 4. Lunch vs Average Score
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='lunch', y=df[['math score', 'reading score', 'writing score']].mean(axis=1),
                data=df, palette='autumn')
    plt.title('Average Scores by Lunch Type')
    plt.xlabel('Lunch')
    plt.ylabel('Average Score')
    plt.tight_layout()
    plt.show()

    # 5. Test Preparation Course vs Average Score
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='test preparation course', y=df[['math score', 'reading score', 'writing score']].mean(axis=1),
                data=df, palette='spring')
    plt.title('Average Scores by Test Preparation Course')
    plt.xlabel('Test Preparation Course')
    plt.ylabel('Average Score')
    plt.tight_layout()
    plt.show()

# Panggil fungsi untuk membuat visualisasi
plot_score_distribution()
````

**Insight :**
Visualisasi menunjukkan bahwa siswa perempuan memiliki nilai lebih tinggi daripada laki-laki di semua mata pelajaran, ada variasi nilai berdasarkan ras/etnisitas, nilai rata-rata meningkat dengan pendidikan orang tua yang lebih tinggi, siswa dengan makanan standard memiliki skor lebih baik, dan mereka yang mengikuti kursus persiapan ujian cenderung berperforma lebih baik.

**Menggunakan Histogram**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca dataset
df = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')

# Set tema untuk seaborn agar tampilan lebih estetik
sns.set_theme(style="whitegrid")

# Membuat histogram untuk distribusi nilai
def plot_histograms():
    plt.figure(figsize=(15, 6))

    # Distribusi Math Score
    plt.subplot(1, 3, 1)
    sns.histplot(df['math score'], bins=20, kde=True, color='skyblue')
    plt.title('Distribution of Math Scores')
    plt.xlabel('Math Score')
    plt.ylabel('Frequency')

    # Distribusi Reading Score
    plt.subplot(1, 3, 2)
    sns.histplot(df['reading score'], bins=20, kde=True, color='lightcoral')
    plt.title('Distribution of Reading Scores')
    plt.xlabel('Reading Score')
    plt.ylabel('Frequency')

    # Distribusi Writing Score
    plt.subplot(1, 3, 3)
    sns.histplot(df['writing score'], bins=20, kde=True, color='lightgreen')
    plt.title('Distribution of Writing Scores')
    plt.xlabel('Writing Score')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Panggil fungsi untuk membuat histogram
plot_histograms()
````

**Insight :**
Secara umum, ketiga distribusi skor (Matematika, Membaca, dan Menulis) menunjukkan pola distribusi normal, di mana sebagian besar siswa mendapatkan skor di sekitar nilai rata-rata.

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')

# Tentukan batas nilai
passing_score = 75

# Buat fungsi untuk menghitung jumlah siswa yang lulus dan gagal
def count_pass_fail(df, subject):
    pass_count = (df[subject] >= passing_score).sum()
    fail_count = (df[subject] < passing_score).sum()
    return pass_count, fail_count

# Hitung jumlah lulus dan gagal untuk setiap pelajaran
math_pass, math_fail = count_pass_fail(df, 'math score')
reading_pass, reading_fail = count_pass_fail(df, 'reading score')
writing_pass, writing_fail = count_pass_fail(df, 'writing score')

# Tampilkan hasil
results = {
    'Subject': ['Math', 'Reading', 'Writing'],
    'Pass': [math_pass, reading_pass, writing_pass],
    'Fail': [math_fail, reading_fail, writing_fail]
}

results_df = pd.DataFrame(results)
print(results_df)
````

**Insight :**
Jumlah siswa yang gagal jauh lebih banyak dibandingkan yang lulus di semua mata pelajaran, dengan matematika memiliki jumlah gagal tertinggi (705), diikuti oleh menulis (650) dan membaca (630), mengindikasikan perlunya intervensi pendidikan untuk meningkatkan kinerja siswa dalam mata pelajaran ini.

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')

# Hitung nilai rata-rata untuk setiap siswa
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Lihat korelasi antara nilai rata-rata dan kategori
categories = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Visualisasikan nilai rata-rata berdasarkan kategori
plt.figure(figsize=(15, 10))

for i, category in enumerate(categories):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x=category, y='average_score', data=df)
    plt.title(f'Average Score by {category}')
    plt.ylabel('Average Score')
    plt.xlabel(category)

plt.tight_layout()
plt.show()

# Hitung dan tampilkan korelasi hanya untuk kolom numerik
numeric_columns = df[['math score', 'reading score', 'writing score', 'average_score']]
correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
````

**Insight :**
Grafik boxplot menunjukkan perbedaan signifikan dalam skor rata-rata berdasarkan gender, ras/etnis, pendidikan orang tua, jenis makan siang, dan keikutsertaan dalam kursus persiapan ujian.

Selain itu, terdapat korelasi kuat antara nilai Matematika, Membaca, Menulis, dan nilai rata-rata, yang menunjukkan bahwa siswa yang berprestasi tinggi di satu mata pelajaran cenderung juga berprestasi tinggi di yang lain.
