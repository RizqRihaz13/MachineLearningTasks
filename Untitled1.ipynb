###**1**. **Memuat** **dan** **Membersihkan** **Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm

# Nama kolom sesuai dengan dataset adult
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]

# Memuat dataset
data_path = 'adult.data'
df = pd.read_csv(data_path, names=column_names, sep=r'\s*,\s*', engine='python')
```

```python
# Cek data apakah ada yang hilang atau tidak konsisten
print(df.info())
print(df.isnull().sum())
```

```python
# Menghapus baris yang memiliki missing value
df = df.replace('?', pd.NA)
df = df.dropna()
```

```python
# Melihat 5 baris pertama
print(df.head())
```

###**2**. **Visualisasi** **Variabel**

*   **Pie** **Chart** **:**

    Visualisasi untuk melihat proporsi jenis kelamin.


```python
# Pie Chart untuk distribusi jenis kelamin
sex_counts = df['sex'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Distribusi Jenis Kelamin')
plt.axis('equal')  # Untuk membuat pie chart bulat
plt.show()
```

**Analisis :**

**Berdasarkan visualisasi tersebut, mahasiswa laki-laki memiliki persentase yang lebih tinggi dibandingkan mahasiswa perempuan. ini menunjukkan bahwa terdapat ketidakseimbangan gender dalam populasi mahasiswa.**

*   **Bar Chart** **:**

    Visualisasi jumlah individu berdasarkan tingkat pendidikan.

```python
# Bar Chart untuk distribusi tingkat pendidikan
education_counts = df['education'].value_counts()

plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='skyblue')
plt.title('Distribusi Tingkat Pendidikan')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Analisis :**

**Berdasarkan** **diagram batang di atas, bahwa sebagian besar individu memiliki pendidikan hingga tingkat SMA (HS-grad) dan perguruan tinggi (Some-college, Bachelors).**



*   **Line Graphs :**

    Melihat hubungan antara usia dan jam kerja per minggu.



```python
# Line Graphs: Usia vs Jam Kerja per Minggu
df_sorted = df.sort_values(by='age')
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['age'], df_sorted['hours-per-week'], color='purple')
plt.title('Usia vs Jam Kerja per Minggu')
plt.xlabel('Usia')
plt.ylabel('Jam Kerja per Minggu')
plt.show()
```



*   **Scatter Plot :**

     Melihat korelasi antara usia dan penghasilan modal.



```python
# Scatter Plot: Usia vs Capital Gain
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['capital-gain'], alpha=0.5, color='green')
plt.title('Scatter Plot Usia vs Capital Gain')
plt.xlabel('Usia')
plt.ylabel('Capital Gain')
plt.show()
```



*   **Heatmap :**

      Menunjukkan korelasi antar variabel numerik.



```python
# Heatmap: Korelasi antar variabel numerik
# Memilih hanya kolom numerik
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Menghitung korelasi antar kolom numerik
corr = df[numerical_columns].corr()

# Membuat Heatmap Korelasi
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```



*   **Grouping (Pivot):**
  
    Mengelompokkan data berdasarkan pekerjaan dan jenis kelamin, lalu menghitung rata-rata jam kerja.

```python
# Grouping: Rata-rata jam kerja berdasarkan pekerjaan dan jenis kelamin
grouped_data = df.pivot_table(values='hours-per-week', index='occupation', columns='sex', aggfunc='mean')
print(grouped_data)

# Visualisasi pivot table
grouped_data.plot(kind='bar', figsize=(10, 6))
plt.title('Rata-rata Jam Kerja berdasarkan Pekerjaan dan Jenis Kelamin')
plt.ylabel('Jam Kerja per Minggu')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```



*   **ANOVA** **:**

    Melakukan uji apakah ada perbedaan signifikan dalam jam kerja antara jenis kelamin

```python
# ANOVA: Uji apakah ada perbedaan signifikan dalam jam kerja antara jenis kelamin
male_hours = df[df['sex'] == 'Male']['hours-per-week']
female_hours = df[df['sex'] == 'Female']['hours-per-week']

anova_result = stats.f_oneway(male_hours, female_hours)
print('Hasil ANOVA:', anova_result)

# Interpretasi hasil ANOVA
if anova_result.pvalue < 0.05:
    print("Ada perbedaan signifikan dalam jam kerja antara pria dan wanita.")
else:
    print("Tidak ada perbedaan signifikan dalam jam kerja antara pria dan wanita.")
```
